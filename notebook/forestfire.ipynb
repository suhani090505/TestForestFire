```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
df.head(5)
```

```python
df.tail(5)
```

```python
df.info()
```

```python
##Data cleaning process
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

The dataset is converted into two sets based on region from 122 index, we make a new column

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head(5)
```

```python
##Remove the null values
df=df.dropna().reset_index(drop=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
##Fix spaces in column name
df.columns=df.columns.str.strip()
df.columns
```

```python
##Change the required columns as integer data type
## day,month,year,temp,rh,ws
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
df.info()
```

```python
df[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']]=df[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']].astype(float)
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

EXPLORATORY DATA ANALYSIS


```python
##Drop day,month and year
df_copy=df.drop(['day','month','year'],axis=1)
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
##Encoding of the categories in classes
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
##Plot density features for all features
import seaborn as sns
%matplotlib inline
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
##Percentage for pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
##Plotting the pie chart
classLabels=["Fire","Not Fire"]
plt.figure(figsize=(20,15))
plt.pie(percentage,labels=classLabels,autopct='%1.1f%%')
plt.show()
```

```python
df_copy.corr()
```

```python
sns.heatmap(df_copy.corr())
plt.show()
```

```python
##Box plots
sns.boxplot(df['FWI'])
plt.show()
```

```python
##Montly Fire Analysis
dftemp=df.loc[df['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.title('Fire analysis of bejaia Regions')
plt.show()
```

```python
dftemp=df.loc[df['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df)
plt.title('Fire analysis of Sidi-Bel Abbes Regions')
plt.show()
```
